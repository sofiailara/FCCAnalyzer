```python
h_mumu = f['p8_ee_Zmumu_ecm91/cosThetac'].to_hist()

w = h_mumu.axes[0].centers

xsec_F = h_mumu[0.0j:1.0j].sum()
print(xsec_F)
xsec_B = h_mumu[-1.0j:0.0j].sum()
print(xsec_B)

AFB=(xsec_F.value-xsec_B.value)/(xsec_F.value+xsec_B.value)
print(AFB)



variance = (4*xsec_B.value*xsec_F.value)/((xsec_B.value+xsec_F.value)**3)
uncertainty = np.sqrt(variance)
print(uncertainty)

rel_uncertainty = uncertainty/AFB
print(rel_uncertainty)

```
